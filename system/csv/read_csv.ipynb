```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../motivation/data.csv')
read = df['CISO_March'].dropna().to_list()
len(read)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=1, gridspec_kw={'hspace': 0.4, 'wspace': 0.2, 'bottom': 0.2, 
                    'top': 0.8, 'right':0.995, 'left':0.17}, figsize=(10,2.5), sharey=False)
axs.plot(read)
```

```python
span = 48
start = 620
fig, axs = plt.subplots(nrows=1, ncols=1, gridspec_kw={'hspace': 0.4, 'wspace': 0.2, 'bottom': 0.2, 
                    'top': 0.8, 'right':0.995, 'left':0.17}, figsize=(6,2), sharey=False)
axs.plot(read[start:start+span+1], marker='o')
axs.set_ylim(50, 350)
# read[start:start+span+1].to_csv('carbon_intensity.csv', index=False)
df['CISO_March'].dropna()[start:start+span+1].to_csv('carbon_intensity.csv', index=False)
```
